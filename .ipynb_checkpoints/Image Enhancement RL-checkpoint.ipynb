```python
# Imports
import numpy as np
from scipy.ndimage.filters import median_filter
import matplotlib.pyplot as plt
import cv2
import time
```

```python
def get_new_state(cur_img, action):

    if action == "w":
        # Weiner filter
        pass
    elif action == "m":
        # Median filter 3 x 3
        return median_filter(image, size=(3,3))
    elif action == "s":
        # Sharpening filter
        kernel = np.array([[0,-1,0], [-1,5,-1], [0,-1,0]])
        return cv2.filter2D(cur_img, -1, kernel)
    else:
        return np.array([])
```

```python
Q = {}
img_path = "41.jpg"
original_image = cv2.imread(img_path, 0)
image = original_image

for i in range(1,10):
    for j in range(1,10):
        
        median = 0.1 * i
        sharp  = 0.1 * j
        image = median*get_new_state(original_image, "m") + sharp*get_new_state(original_image, "s")
        
        print(" Iteration : ", i, " , ", j)
        print(" #### Original ####")
        plt.imshow(original_image)
        plt.show()
        
        print(" #### Transformed ####")
        plt.imshow(image)
        plt.show()
        
#         reward = float(input(" Enter reward : "))
#         img_tuple = (tuple(x) for x in image)
#         Q[img_tuple] = reward
```
